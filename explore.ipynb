```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('data/raw/shopping.csv')
df.head()
df.info()
df.describe()
```

```python
df
```

```python
# 1. Which clothing items are purchased most frequently by gender?
df.pivot_table(
    index='Item Purchased',
    columns='Gender',
    aggfunc='size',
    fill_value=0
).sort_values(by=['Female','Male'], ascending=False).head(10)
```

The shorts is the clothing item that was purchased the most frequently. Female customers had the highest amount of shorts purchased compared to males. They purchased 7 more than males in this dataset.

```python
# 2. How does purchasing frequency relate to whether customers leave a review and to the rating they give?
df['Left Review'] = df['Review Rating'].notnull()
df.groupby('Left Review')['Review Rating'].value_counts()
```

```python
df.groupby(['Left Review', 'Review Rating'])['Frequency of Purchases'].value_counts()
```

Most customers who left a review gave a 4‑star rating, making it the most common score with 715 total responses. Among those 4‑star reviewers, the largest group consisted of customers who typically make purchases on a quarterly basis, accounting for 133 customers.

```python
# 3. Does the seasonality of an item influence the purchase amount customers spend?
df.groupby('Season')['Purchase Amount (USD)'].mean()
```

Spring show the highest average purchase amount at $39.29 USD. However, seasonal differences are minimal, with average spending ranging from $38.37 to $39.29 USD.

```python
# 4. Which states generate the highest sales volume?
df.groupby('Location')['Purchase Amount (USD)'].sum().sort_values(ascending=False).head(10)
```

The top three states with the highest sales volume are California, Nebraska, and Delaware, with sales total of $3474.55, $3195.73, and $3140.07 USD respectively. 

```python
# 5. What are the most popular colors by season?
df.groupby(['Season'])['Color'].value_counts().groupby('Season').idxmax()
```

In Fall, the most popular color is Brown. In Spring, the most popular color is Baby Blue. In Summer, the most popular color is Lavender. In Winter, the most popular color is Black.

```python
# 6. What is the most popular clothing item by season?
df.groupby(['Season'])['Item Purchased'].value_counts().groupby('Season').idxmax()
```

In Fall, Spring, Summer, and Winter, the most popular clothing items are socks, running shoes, shorts, and leggings, respectively. Fall and Spring share a preference for footwear, while Summer and Winter show stronger popularity for lower‑body apparel.

```python
# 7. What is the effect of promo codes on the dollar amount of purchases?
df.groupby('Promo Code Used')['Purchase Amount (USD)'].mean()
```

Customers who use promo codes tend to spend more on average than those who do not. The difference in the average purchase amount between the two groups is $19.85 USD, indicating that promo‑code users make higher‑value purchases overall.

```python
# 8. When do users leave a review?
df['Left Review'] = df['Review Rating'].notnull()
df.groupby(['Season', 'Left Review']).size().unstack()
```

The data shows that more customers chose not to submit a rating review than those who did. For example, in the Fall, 669 customers did not leave a review compared with 387 who did. Among those who did submit a review, Winter had the highest participation with 499 reviews, while Spring had the lowest with 205.
